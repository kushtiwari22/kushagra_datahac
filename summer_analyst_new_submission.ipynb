```python
import pandas as pd
import numpy as np
```

```python
train=pd .read_csv("/kaggle/input/summer-analyst/training_set_features.csv")
train_label=pd.read_csv("/kaggle/input/summer-analyst/training_set_labels.csv")
test=pd.read_csv("/kaggle/input/summer-analyst/test_set_features.csv")
```

```python
# columns for backfill
columns1=['respondent_id', 'xyz_concern', 'xyz_knowledge',
       'behavioral_antiviral_meds', 'behavioral_avoidance',
       'behavioral_face_mask', 'behavioral_wash_hands',
       'behavioral_large_gatherings', 'behavioral_outside_home',
       'behavioral_touch_face', 'doctor_recc_xyz', 'doctor_recc_seasonal',
       'chronic_med_condition', 'child_under_6_months', 'health_worker', 'opinion_xyz_vacc_effective', 'opinion_xyz_risk',
       'opinion_xyz_sick_from_vacc', 'opinion_seas_vacc_effective',
       'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'age_group',
       'education', 'race', 'sex', 'hhs_geo_region', 'census_msa',
       'household_adults', 'household_children']

# columns to fill  dont_know
columns2=['employment_industry','employment_occupation','health_insurance','income_poverty', 'marital_status','rent_or_own', 'employment_status']
```

```python
train[columns1]=train[columns1].fillna(method='ffill')
train[columns2]=train[columns2].fillna('dont_know') # method1
# train=train.drop(columns=columns2) # method2
test[columns1]=test[columns1].fillna(method='ffill')
test[columns2]=test[columns2].fillna('dont_know') # method1
```

```python
train.isnull().sum()
test.isnull().sum()
```

```python
columns3 = train.select_dtypes(include='object').columns
```

```python
train = pd.get_dummies(train , columns=columns3)
test = pd.get_dummies(test , columns=columns3)
train.shape
```

```python
train_label=train_label.drop(columns=["respondent_id"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import xgboost as xgb
x_train , x_test ,y_train ,  y_test= train_test_split(train , train_label ,test_size=0.2 , random_state=42 )
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train.columns = [col.replace('[', '').replace(']', '').replace('<', '').replace(' ', '_') for col in x_train.columns]
test.columns = [col.replace('[', '').replace(']', '').replace('<', '').replace(' ', '_') for col in test.columns]
```

```python
x_test.columns = [col.replace('[', '').replace(']', '').replace('<', '').replace(' ', '_') for col in x_test.columns]
# x_test.columns = [col.replace('[', '').replace(']', '').replace('<', '').replace(' ', '_') for col in x_test.columns]
```

```python
y_test.columns = [col.replace('[', '').replace(']', '').replace('<', '').replace(' ', '_') for col in y_test.columns]
# y_test.columns = [col.replace('[', '').replace(']', '').replace('<', '').replace(' ', '_') for col in y_test.columns]
```

```python
y_train.columns = [col.replace('[', '').replace(']', '').replace('<', '').replace(' ', '_') for col in y_train.columns]
# y_train.columns = [col.replace('[', '').replace(']', '').replace('<', '').replace(' ', '_') for col in y_train.columns]
```

```python
x_train.shape
```

# **random_forest**

```python
random_forest=RandomForestClassifier(n_estimators=150, random_state=42)
random_forest.fit(x_train , y_train)
y_predict1= random_forest.predict_proba(test)
```

```python
predict1=np.array(y_predict1)
class1_prob=predict1[0]
class2_prob=predict1[1]
class1_prob_rf=class1_prob[:,1]
class2_prob_rf=class2_prob[:,1]
```

# **xgboost**

```python
xboost=xgb.XGBClassifier(n_estimators=100, eval_metric='logloss', random_state=42)
xboost.fit(x_train , y_train)
y_predict2 = xboost.predict_proba(test)
```

```python
predict2=np.array(y_predict2)
```

```python
class1_prob_xg=predict2[:,0]
class2_prob_xg=predict2[:,1]
```

```python
class2_prob_xg.shape
```

```python
sub_csv=pd.read_csv("/kaggle/input/summer-analyst2/submission_format.csv")
sub_csv.head()
```
